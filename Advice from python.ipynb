```python
import time
import os
import time
import random
try:
    from pynput.keyboard import Key, Controller
except:
    os.system('pip install pynput')
    from pynput.keyboard import Key, Controller


x='''
import random
from  this import s,d
import sys
import os


try:
    
    from win10toast import ToastNotifier

except:
    os.system('pip install win10toast')
    from win10toast import ToastNotifier
  

def give_me_advice(d=d,s=s,r=random.randint(2,20)):
    return "".join([d.get(c, c) for c in s]).split('\\n')[r]


ToastNotifier().show_toast("My advice to you is : ", give_me_advice(), duration = 10, threaded = True)
'''



f = open("advice.py", "w")


f.write(x)
f.close()

keyboard = Controller()

os.system('control schedtasks')
time.sleep(5)
keyboard.press(Key.alt)
keyboard.type("ab")

time.sleep(2)
keyboard.release(Key.alt)
time.sleep(1)
keyboard.type("daily advice")
for i in range(4):
    keyboard.press(Key.enter)
    time.sleep(1)
keyboard.type(f"{sys.executable}")
time.sleep(1)
keyboard.press(Key.tab)
time.sleep(1)
keyboard.press(Key.tab)
time.sleep(1)
keyboard.type("advice.py")
time.sleep(1)
keyboard.press(Key.tab)
keyboard.type(f"{os.getcwd()}")
for i in range(2):
    keyboard.press(Key.enter)
    time.sleep(1)


exec(x)


```
